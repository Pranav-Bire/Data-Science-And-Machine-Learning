```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
points = np.array([[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85],
                   [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]])
```

```python
points.shape
```

```python
i_centroids = np.array([[0.1, 0.6], [0.3, 0.2]])
```

```python
kmeans = KMeans(n_clusters=2, init=i_centroids, n_init=1, random_state=0).fit(points)
```

```python
c_assign = kmeans.labels_
```

```python
c_centers = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(points[c_assign == 0, 0], points[c_assign == 0, 1], color='red', label='Cluster 1')
```

```python
plt.scatter(points[c_assign == 1, 0], points[c_assign== 1, 1], color='blue', label='Cluster 2')
```

```python
plt.scatter(c_centers[:, 0], c_centers[:, 1], color='black', marker='x', s=100, label='Centroids')
```

```python
P6 = np.array([0.25, 0.5])
plt.scatter(P6[0], P6[1], color='green', marker='o', s=100, label='P6')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('K-Means Clustering')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# a) Which cluster does P6 belong to?

cluster_P6 = kmeans.predict(P6.reshape(1, -1))
print("a) P6 belongs to Cluster:", cluster_P6[0]+1)  # Adding 1 to convert 0-indexed cluster labels to 1-indexed
```

```python
# b) What is the population of a cluster around m2?
pop_cluster2 = np.sum(c_assign== 1)
print("b) Population of the cluster around m2 (C2):", pop_cluster2)
```

```python
# c) What is the updated value of m1 and m2?
print("c) Updated value of m1:", c_centers[0])
print("Updated value of m2:", c_centers[1])
```
