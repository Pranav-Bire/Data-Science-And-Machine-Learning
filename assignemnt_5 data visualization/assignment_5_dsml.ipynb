```python
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
print(sns.get_dataset_names())
```

```python
file=sns.load_dataset("car_crashes")
print(file.head()
)
```

```python

sns.relplot(data=file,x="speeding",y="no_previous",hue="abbrev", palette="bright")
plt.title("speeding vs previous crash", fontsize=15)
sns.set_style("white")
sns.set_style("darkgrid",{'grid.color':'.5'})
sns.despine()
plt.show()
```

```python
sns.countplot(y="no_previous", palette="bright" ,data=file)
plt.title("speeding vs previous crash", fontsize=15)


plt.legend(facecolor="blue", bbox_to_anchor=(1,1))
plt.show()
```

```python
sns.barplot(x = 'no_previous', y = 'speeding', data = file,
            palette = 'PuRd',ci=None )
plt.legend()
plt.show()
print(file.columns)
```

```python
plt.hist(file.speeding, color="red",edgecolor="black" )
plt.legend()
plt.show()
```

```python
sns.boxplot(x="total", y="speeding", data=file)
```

```python
sns.jointplot(x="alcohol", y="total",
                   data=file, palette="Accent" , kind='kde',color="black")
```

```python
plt.stem(file.alcohol,file.total, linefmt=":" , markerfmt="r", bottom=2 , basefmt="green", label="speeding")
orientation="horizontal"
plt.legend()
plt.show()
```
